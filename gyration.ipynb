```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


amber_gyr_pairs = pd.read_csv('native_TTR/gyrate_amber.txt', sep = '\\s+', header=None)
amber_gyr_pairs.columns = ['time', 'gyr', 'g', 'h', 'j']
amber_gyr_pairs.drop(['g', 'h', 'j'], axis=1, inplace = True)
amber_gyr_pairs['time'] = amber_gyr_pairs['time']/1000

smogtogromos_gyr_pairs = pd.read_csv('native_TTR/gyrate_smogtogromos.txt', sep = '\\s+', header=None)
smogtogromos_gyr_pairs.columns = ['time', 'gyr', 'g', 'h', 'j']
smogtogromos_gyr_pairs.drop(['g', 'h', 'j'], axis=1, inplace = True)
smogtogromos_gyr_pairs['time'] = smogtogromos_gyr_pairs['time']/1000


#greta_gyr_pairs = pd.read_csv('native_TTR/gyrate_greta.txt', sep = '\\s+', header=None)
#greta_gyr_pairs.columns = ['time', 'gyr', 'g', 'h', 'j']
#greta_gyr_pairs.drop(['g', 'h', 'j'], axis=1, inplace = True)
#greta_gyr_pairs['time'] = greta_gyr_pairs['time']/1000


greta2_gyr_pairs = pd.read_csv('native_TTR/gyrate_greta_exex_harp0.txt', sep = '\\s+', header=None)
greta2_gyr_pairs.columns = ['time', 'gyr', 'g', 'h', 'j']
greta2_gyr_pairs.drop(['g', 'h', 'j'], axis=1, inplace = True)
greta2_gyr_pairs['time'] = greta2_gyr_pairs['time']/1000
```

```python
fig, axes = plt.subplots(1,2, figsize = (10,5), sharey=True)

axes[0].set_ylabel("Gyration radious")
axes[0].set_xlabel("Time (ns)")

sns.lineplot(ax=axes[0], data=smogtogromos_gyr_pairs, y="gyr", x='time', color= '#985098', alpha=0.5)
#sns.lineplot(ax=axes[0], data=greta_gyr_pairs, y="gyr", x='time', color= '#989850', alpha=0.5)
sns.lineplot(ax=axes[0], data=amber_gyr_pairs, y="gyr", x='time', color= '#509898', alpha=0.5)

sns.histplot(ax=axes[1], data=smogtogromos_gyr_pairs, y = 'gyr', stat = 'density', binwidth = 0.02, binrange=[0.5,1.3],kde=True, color='#985098')
sns.histplot(ax=axes[1], data=amber_gyr_pairs, y = 'gyr', stat = 'density', binwidth = 0.02, binrange=[0.5,1.3],kde=True, color='#509898')
#sns.histplot(ax=axes[1], data=greta_gyr_pairs, y = 'gyr', stat = 'density', binwidth = 0.02, binrange=[0.5,1.3],kde=True, color='#989850')

fig.tight_layout()
```
